##### Q1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
import logging
logging.basicConfig(filename="Bank_Account",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')

class bankaccount:
    logging.info("Inside the class bankaccount")
    
    def __init__(self,account_no,account_holder_nm,balance):
        logging.info("Initialisation of the attributes within the __init__() method")
        
        self.account_no = account_no
        logging.info("Assigning the account number to the variable account_no")
        self.account_holder_nm = account_holder_nm
        logging.info("Assigning the Account Holder's name")
        self.balance = balance
        logging.info("Total balance in the account")

    def deposit_amount(self,amount):
        logging.info("Inside the deposit_amount method")
        self.balance += amount
        logging.info("Credited the amount in the account balance")
        print("Deposit amount:",self.balance)
        logging.info("Checking the total balance in the account")

    def withdraw_amount(self,amount):
        logging.info("Inside the withdraw_amount method")
        self.balance -= amount
        logging.info("Debited the amount from the account balance")
        print("Total Amount Withdraw:",self.balance)
        logging.info("Checking the total balance in the account")

    def display_balance(self):
        logging.info("Inside the display_balance method")
        print("Account Number:",self.account_no)
        logging.info("Displaying the account number of an User")
        print("Account Holder Name:",self.account_holder_nm)
        logging.info("Displaying the account holder's name")
        print("Total Balance:",self.balance)
        logging.info("Displaying the total balance in the account")

account1=bankaccount(1234567,"Kajal Bharti",5000)
logging.info("Creating an object or instance for the class bankaccount")

account1.deposit_amount(2000)
logging.info("Printing the total amount credited in the account.")
account1.withdraw_amount(4000)
logging.info("Printing the total amount debited from the account")
account1.display_balance()
logging.info("Displaying the details of bankaccount")
```

```python
import logging
logging.basicConfig(filename="Vehicle_Rental_System",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')

class vehicle_Rental_system:

    logging.info("Inside the class vehicle_Rental_system")
    def __init__(self,vehicle_avail):
        logging .info("Inside the __init__() method")
        self.vehicle_avail = {}
        logging.info("created an attribute vehicle_avail")

    def rent_vehicle(self, vehicle_type, vehicle_id):
        logging.info("Inside the rent_vehicle() function")

        if vehicle_type in self.vehicle_avail and vehicle_id in self.vehicle_avail[vehicle_type]:
            logging.info("Checking if the vehicle is available or not")
            self.vehicle_avail[vehicle_type].remove(vehicle_id)
            logging.info("If Vehicle is available then remove the vehicle_id from the list as we have rented the vehicle")
            print("Rented {} with {}".format(vehicle_type, vehicle_id))
            logging.info("Printing that the vehicle is on rent")
        else:
            logging.info("Inside the else block.")
            print("{} with {} is not available".format(vehicle_type, vehicle_id))
            logging.info("If vehicle is not available then this block will execute.")

    def return_vehicle(self,vehicle_type, vehicle_id):
        logging.info("Inside the return_vehicle function")

        if vehicle_type in self.vehicle_avail :
            logging.info("Checking if vehical_type is available in vehicle_avail")
            self.vehicle_avail[vehicle_type].append(vehicle_id)
            logging.info("We are appending the vehicle_id in the vehicle_avail as we have returned the vehicle")
            print("Returned {} wth {}".format(vehicle_type, vehicle_id))
            logging.info("Printing that the vehicle has been returned.")

    def available_vehicle(self) :
        logging.info("Inside the available_vehicles Function.")

        for vehicle_type , vehicle_id in self.vehicle_avail.items():
            logging.info("Inside the for loop.")
            print(f"Available {vehicle_type}: {','.join(vehicle_id)}")
            logging.info("printing the available vehicles")

vehicle_rental=vehicle_Rental_system({})
logging.info("creating an object or an instance of the class vehicle_Rental_system")

vehicle_rental.vehicle_avail["car"]=['car101','car102','car103']
logging.info("Added the available vehicles")
vehicle_rental.vehicle_avail["bike"]=['bike101','bike102']
logging.info("Added the available vehicles")

logging.info("Displaying the rented and returned vehicles.")
vehicle_rental.rent_vehicle("car",'car101')
vehicle_rental.rent_vehicle("bike",'bike101')
vehicle_rental.rent_vehicle("scooty",'scooty102')
vehicle_rental.return_vehicle("car",'car101')

logging.info("Displaying the available vehicles.")
vehicle_rental.available_vehicle()
```

##### Q4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
import logging
logging.basicConfig(filename="Library_Catalog_System",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')

class book:
    logging.info("Inside the class book")

    def __init__(self,title,author):
        logging.info("Inside the constructor block")
        self.title = title
        self.author = author
        self.available=True

class Library(book):
    logging.info("Inside the class Library")

    def __init__(self):
        logging.info("Inside the constructor block")
        self.books =[]

    def add_book(self,title,author):
        logging.info("Inside the add_book method block")
        Book=book(title,author)
        self.books.append(Book)
        logging.info("We are trying to append the book details.")
        print("Book {} by {} added to the library.".format(title,author))

    def display_books(self):
        logging.info("Inside the disply_books method block.")
        if not self.books:
            logging.info("Inside the if else block to check if the book is available.")
            print("No books are available in the library.")
        else:
            print("Available books in the library are")
            for book in self.books:
                logging.info("If the book is available then it will return the available books.")
                if book.available:
                    return (book)
                
    def borrow_book(self,title):
        logging.info("Inside the borrow_book method block.")
        for book in self.books:
            if book.title==title and book.available:
                logging.info("Inside the if else block to check if book is available then we can borrow it.")
                book.available=False
                print("You have borrowed book {} by {}.".format(title,book.author))
            else:
                return ("Sorry! book is not available or has been already borrowed.")
            
library=Library()
logging.info("LibraryCreating an object or an instance of class Library")

logging.info("Adding the books in library")
library.add_book("The Great Gatsby", "F.Scott Fitzgerald")
library.add_book("To Kill a Mockingbird", "Harper Lee")
library.add_book("1984", "George Orwell")

logging.info("Displaying the available books in the library.")
library.display_books()

logging.info("Displaying the book that has been borrowed")
library.borrow_book("The Great Gatsby")
```

##### Q5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
import logging
logging.basicConfig(filename="Product_Inventory",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')

class Product:
    logging.info("Inside the class Product.")
    def __init__(self, product_id, name, price, quantity):
        logging.info("Inside the constructor block.")
        logging.info("Initialising the attributes.")
        self.product_id = product_id  
        self.name = name  
        self.price = price  
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        logging.info("Inside the update_quantity method block")
        self.quantity = new_quantity

    def __str__(self):
        logging.info("Inside the __str__ method.")
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: {self.price} USD, Quantity: {self.quantity}"

class Inventory:
    logging.info("Inside the class Inventory")
    
    def __init__(self):
        logging.info("Inside the consturctor block.")
        self.products = []  

    def add_product(self, product):
        logging.info("Inside the add_products method block to add the product.")
        self.products.append(product)

    def update_product_quantity(self, product_id, new_quantity):
        logging.info("Inside the update_product_quantity method block to update the product quantity details.")
        for product in self.products:
            if product.product_id == product_id:
                product.update_quantity(new_quantity)

    def display_available_products(self):
        logging.info("Inside the display_available_products method to display the details of available products.")
        print("Available Products in the Inventory:")
        for product in self.products:
            print(product)

inventory = Inventory()
logging.info("Creating an instance or object for the class Inventory")

logging.info("Creating an instance or object for the class Product")
product1 = Product(1, "Laptop", 1000, 10)
product2 = Product(2, "Smartphone", 500, 20)

logging.info("Adding the products")
inventory.add_product(product1)
inventory.add_product(product2)

logging.info("Updating the product details.")
inventory.update_product_quantity(1, 8)

logging.info("Displaying the available products.")
inventory.display_available_products()
```

##### Q6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
import logging
logging.basicConfig(filename="Shape_Calculation",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')

class Shape:
    logging.info("Inside the class Shape.")

    def __init__(self, length, width=None, height=None):
        logging.info("Inside the __init__ method.")
        logging.info("Initialising the attributes.")
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        logging.info("Inside the calculate_area method to calculate the area and if not then it will rais an exception.")
        raise NotImplementedError("Subclasses must implement this method")

    def calculate_perimeter(self):
        logging.info("Inside the calculate_perimeter method to calculate the perimeter and if not then it will rais an exception.")
        raise NotImplementedError("Subclasses must implement this method")

class Rectangle(Shape):
    logging.info("Inside the class Rectangle which is inherited from the parent class.")
    
    def calculate_area(self):
        logging.info("Inside the calculate_area method to calculate the area of the rectangle.") 
        if self.width is None:
            return self.length * self.length  
        else:
            return self.length * self.width

    def calculate_perimeter(self):
        logging.info("Inside the calculate_perimeter method to calculate the perimeter of a rectangle.")
        if self.width is None:
            return 4 * self.length  
        else:
            return 2 * (self.length + self.width)

class Circle(Shape):
    logging.info("Inside the class Circle which is inherited from the parent class Shape.")

    def calculate_area(self):
        logging.info("Inside the calculate_area method to calculate the area of the circle")
        return 3.14159265359 * self.length**2

    def calculate_perimeter(self):
        logging.info("Inside the calculate_perimeter method to calculate the perimeter of the circle")
        return 2 * 3.14159265359 * self.length

class Triangle(Shape):
    logging.info("Inside the class Triangle which is inherited from the parent class Shape.")

    def calculate_area(self):
        logging.info("Inside the calculate_area method to calculate the area of the triangle")
        if self.width is None or self.height is None:
            logging.info("Inside the if block to check if self.width and self.height is none then it will raise an exception.")
            raise ValueError("Both width and height are required for triangle area calculation")
        return 0.5 * self.length * self.width

    def calculate_perimeter(self):
        logging.info("Inside the calculate_perimeter method to calculate perimeter of triangle.")
        if self.width is None or self.height is None:
            logging.info("Inside the if block to check if self.width and self.height are none then it will raise an exception.")
            raise ValueError("Both width and height are required for triangle perimeter calculation")
        return self.length + self.width + self.height
    
rectangle = Rectangle(5, 4)
logging.info("Creating an instance or object of the class Rectangle.")
circle = Circle(3)
logging.info("Creating an instance or object of the class Circle")
triangle = Triangle(5, 4, 3)
logging.info("Creating an instance or object of the class Triangle.")

logging.info("After calculation , printing the result.")
print("Rectangle - Area:", rectangle.calculate_area(), "Perimeter:", rectangle.calculate_perimeter())
print("Circle - Area:", circle.calculate_area(), "Circumference:", circle.calculate_perimeter())
print("Triangle - Area:", triangle.calculate_area(), "Perimeter:", triangle.calculate_perimeter())
```

```python
import logging
logging.basicConfig(filename="Student_Management_System",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')
class student_management_system:
    logging.info('Inside the class student_management_system')

    def __init__(self,name, student_id):
        logging.info('Inside the __init__() method')
        self.student_id = student_id
        logging.info("Initialised the student_id attribute")
        self.name = name
        logging.info("Initialised the name attribute")
        self.grade=[]
        logging.info("Initialised an empty grade attribute")

    def grading(self,grade):
        logging.info("Inside the grading method")
        self.grade.append(grade)
        logging.info("Adding the grades to the empty grade list")

    def average_grade(self):
        logging.info("Inside the average_grade method")
        if len(self.grade)==0:
            logging.info("Checking if the grades are equal to 0 or not.")
            return 0
            logging.info("If grades are equal to 0 then return 0")
        total =sum(self.grade)
        logging.info("Addition of the grades")
        return total/len(self.grade)
        logging.info("Calculation of the tottal grades.")
    
    def display_stud_details(self):
        logging.info("Inside the display_stud_details method")
        print("student_id=",self.student_id)
        logging.info("Displaying the student's id")
        print("Name=",self.name)
        logging.info("Displaying the student's name")
        print("Grade=",",".join(map(str,self.grade)))
        logging.info("Displaying the grades of the student.")
        print("Total grade=",self.average_grade())
        logging.info("Displaying the total grade of the student")

student1=student_management_system("Rohan",1101)
logging.info("creating an object or an instance of the class student_management_system")
student1.grading(85)
student1.grading(78)
student1.grading(98)

student2=student_management_system("Rohit",1102)
logging.info("creating an object or an instance of the class student_management_system")
student2.grading(88)
student2.grading(75)
student2.grading(95)

student1.display_stud_details()
logging.info("Displaying the details of the first student")

student2.display_stud_details()
logging.info("Displaying the details of the second student")
```

##### Q9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
import logging
logging.basicConfig(filename="Profile",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')
class social_media_profile:
    logging.info("Inside the class social_media_profile")

    def __init__(self,username):
        logging.info("Inside the __init__ method")
        self.username = username
        logging.info("Initialised the username attribute")
        self.posts=[]

    def add_post(self,post):
        logging.info("Inside the add_post method")
        self.posts.append(post)
        logging.info("adding the post")

    def display_post(self):
        logging.info("Inside the display_post method")
        print("posts for {}".format(self.username))
        for post in self.posts:
            print(post)
            logging.info("Displaying the details of the posts")

    def search_posts(self,keywords):
        logging.info("Inside the search_posts method")
        matching_posts = [post for post in self.posts if keywords in post]
        print("Posts containing {} in {}".format(keywords, self.username))
        for post in matching_posts:
            print(post)
            logging.info("Searching the posts using the keywords")

profile=social_media_profile("Kajal")
logging.info("created an object or an instance of class social_media_profile")

profile.add_post("Just had a great day at the park!")
profile.add_post("Trying out a new recipe for dinner tonight.")
profile.add_post("Enjoying the latest book I'm reading.")
logging.info("Adding the posts to the profile")

profile.display_post()
logging.info("Displaying all the posts")

profile.search_posts("great")
logging.info("Displaying the posts by searching using the keywords.")
```

##### Q10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
import logging
logging.basicConfig(filename="ToDoList",level=logging.DEBUG,format='%(asctime)s-%(levelname)s-%(message)s')
class ToDoList:
    logging.info('Inside the class ToDoList')

    def __init__(self):
        logging.info("Inside the __init__ method")
        self.tasks = {}

    def add_tasks(self,task,due_dates):
        logging.info("Inside the add_tasks method")
        self.tasks[task]=due_dates
        logging.info("Adding tasks")

    def tasks_completed(self, task):
        logging.info("Inside the tasks_completed method")
        if task in self.tasks:
            del self.tasks[task]
            logging.info("Deleting the tasks completed")

    def display_pending_tasks(self):
        logging.info("Inside the display_pending_tasks method")
        if not self.tasks:
            logging.info("Checking if any tasks are pending inside the if block")
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print("Task: {}, Due Date: {}".format(task, due_date))
                logging.info("Printing the pending tasks.")

to_do_list = ToDoList()
logging.info("Created an object or an instance of the class ToDoList")

to_do_list.add_tasks("walking","2023-10-25")
to_do_list.add_tasks("swimming","2023-10-27")
to_do_list.add_tasks("Coaching","2023-10-28")
logging.info("Adding the tasks")

to_do_list.display_pending_tasks()
logging.info("Displaying the pending tasks")

to_do_list.tasks_completed("walking")
logging.info("Displaying the tasks completed")

to_do_list.display_pending_tasks()
logging.info("Again displaying the pending tasks after checking the completed tasks")
```
